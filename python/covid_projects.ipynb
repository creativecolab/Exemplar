```python
import spacy
from spacy.lang.en import English
from spacy.lang.en.stop_words import STOP_WORDS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Read data

```python
df_idea = pd.read_json('covid_projects.json');
df_idea.head()
```

```python
check_item = 'description';
drop_list=[];
for i in range(len(df_idea)):
    doc = bert(df_idea[check_item][i]);
    if doc._.language['language'] != 'en':
        drop_list.append(i);
        print(i, doc);
#     else:
#         print(i);
print(drop_list);

#df_idea
```

Drop non-English row from the table and reset index

```python
#print(len(df_idea));
df_idea_clean = df_idea.drop(df_idea.index[drop_list]);
df_idea_clean = df_idea_clean.reset_index(drop=True);
#print(len(df_idea_clean));
```

Text processing
 - remove punctuations, including \r\n  \ '
 - remove stop words

```python
df_idea_clean['description'][70]
```

Compare similarity between ideas

```python
doc_sim = [];
comp_num = 2;
comp_item = 'description';
doc0 = bert(df_idea_clean[comp_item][comp_num]);
print(doc0);

for j in range(len(df_idea_clean)):
    if comp_num != j:
        doc1 = bert(df_idea_clean[comp_item][j]);
        sim = doc0.similarity(doc1);
        doc_sim.append(sim);
        print(j, sim);
```

```python
print(np.mean(doc_sim),
np.max(doc_sim), np.argmax(doc_sim),
np.min(doc_sim), np.argmin(doc_sim),)
```

```python
print('comparing idea:',df_idea['name'][comp_num],df_idea['description'][comp_num])
print('\n')
print('most similar: ',df_idea['name'][np.argmax(doc_sim)],df_idea['description'][np.argmax(doc_sim)])
print('\n')
print('least similar: ',df_idea['name'][np.argmin(doc_sim)],df_idea['description'][np.argmin(doc_sim)])
```

```python
print('Project Name: ',df_idea_clean['name'][comp_num]);
print('Full description: ',df_idea_clean['description'][comp_num]);
print('Short description: ',df_idea_clean['short_description'][comp_num]);
print('Categories: ',df_idea_clean['project_type_list'][comp_num]);

plt.hist(pd.array(doc_sim), bins = 500, range=[0.4,1]);

#print(final_idea[n+2]); # starting from row 2
```
